# Введение в машинное обучение

**Представьте, что у нас есть задача: автоматически сортировать яблоки на конвейере, разделяя их на "хорошие" и "испорченные".**  

Эту задачу можно решить двумя способами: классическим подходом или с использованием машинного обучения.  

<center><img src ="https://ml.gan4x4.ru/msu/dev-2.3/L01/apple_selection.jpg" width="400"></center>


### **Вариант №1. Классический подход**  
Мы можем придумать список правил и эвристик, используя знания о мире. Перенести эти правила в код и получим программу, которая:
1. Получает изображение яблока.  
2. Ищет на нём пятна, повреждения или другие признаки порчи, подсчитывая количество пикселей черного или коричневого цвета.
3. Если таких пикселей больше какого то значение то программа помечает яблоко как "испорченное". Если их нет, оно считается "хорошим".  

На первый взгляд кажется, что это несложно, но на практике возникают проблемы:  
- Пятна на яблоке могут быть разного размера и цвета. Иногда даже хорошие яблоки могут выглядеть "подозрительно" из-за незначительных дефектов.  
- Освещение конвейера может создавать тени или блики, которые похожи на пятна.  
- Испорченное яблоко может быть не только с пятнами, но и, например, с засохшим участком, который не так просто обнаружить.  

Пытаясь исправть эти недостатки мы дописываем правила в код. В итоге программа становится очень сложной: мы добавляем всё больше правил и исключений, но всегда остаются случаи, которые она обрабатывать не умеет.  




### **Вариант №2. Машинное обучение**  
С машинным обучением задача решается по-другому:  
1. Сначала мы собираем данные — фотографируем много яблок на конвейере и отмечаем, какие из них "хорошие", а какие "испорченные".  
2. Загружаем эти данные в модель машинного обучения. Она сама изучит, как выглядят хорошие и испорченные яблоки: какие у них цвета, текстуры, формы и другие особенности. Найдет скрыте закономерности, между "испорченными" и "хорошими" яблоками.
3. После обучения модель сможет автоматически распознавать состояние новых яблок, даже если они имеют необычные дефекты или сняты при разном освещении.  



**Почему машинное обучение лучше в этой задаче?**  
В классическом подходе мы пытаемся вручную прописать все правила, но эта задача становится слишком сложной и затратной.

Машинное обучение позволяет модели изучить закономерности в данных самостоятельно. Ей не нужно объяснять, как выглядят пятна или повреждения — достаточно показать примеры, и она сама поймёт, как их находить.  


Если бы задача была простой (например, просто разделить яблоки по цвету), машинное обучение могло бы показаться избыточным. Но когда задача сложнее и содержит множество вариаций (разные виды дефектов, освещение, смешение факторов), машинное обучение оказывается мощным инструментом, который упрощает разработку и делает решение более универсальным.  


##  AI, ML, DL

**Общепринятые термины в машинном обучении**

<center><img src ="https://ml.gan4x4.ru/msu/dev-2.2/L01/out/ai_ml_dl.png" width="600"></center>

**Искусственный интеллект (AI/ИИ)**  — область IT/Computer science, связанная с моделированием интеллектуальных или творческих видов человеческой деятельности.

**Машинное обучение (ML)** — подраздел ИИ, связанный с разработкой алгоритмов и статистических моделей, которые компьютерные системы используют для выполнения задач без явных инструкций.

**Глубокое обучение (Deep Learning, DL)** — совокупность методов машинного обучения, основанных на искуcственных нейронных сетях и обучении представлениям (**feature/representation learning**). Данный класс методов автоматически выделяет из необработанных данных необходимые признаки (представления), в отличие от методов ML, в которых признаки создают люди вручную (**feature engineering**).

Существует множество определений сильного и слабого ИИ, рассуждений о появлении искусственного сознания и восстании машин.

Всё намного **приземлённее**. Есть набор **объектов $X$** и набор **ответов $Y$**. Пары "объект-ответ" составляют **обучающую выборку**.

Мы будем заниматься **восстановлением решающей функции $F$**, которая переводит признаки $X$, описывающие объекты, в ответы $Y$.

$$ F: X \xrightarrow\ Y $$



# Типы задач в ML

## Базовые задачи

### Классификация

<center><img src="https://ml.gan4x4.ru/msu/dev-2.2/L01/out/classification.png" width="650"></center>

Классификация — **отнесение образца к одному из нескольких попарно не пересекающихся множеств**.

В качестве образцов могут выступать различные по своей природе объекты, например:
* символы текста,
* изображения,
* звуки.

### Регрессия

<center><img src="https://ml.gan4x4.ru/msu/dev-2.2/L01/out/regression.png" width="650"></center>

В случае регрессионных задач у нас, как правило, нет жесткого ограничения на пространство ответов. И мы предсказываем некоторое число, в общем случае вещественное.

### Кластеризация

<center><img src="https://ml.gan4x4.ru/msu/dev-2.2/L01/out/clustering.png" width="650"></center>

Кластеризация — **разбиение множества входных данных на группы**. Ни количество, ни специфические признаки заранее не известны.




Традиционно машинное обучение разделяют на **обучение с учителем** (когда есть ответы к обучающим примерам, например, задача классификации) и **обучение без учителя** (когда ответов/меток нет, например, задача кластеризации)

Существуют различные типы данных с различными типами связности:

* последовательности (важны порядок данных, время),
* пространственно-структурированная информация (изображения, видео),
* табличные данные.



Наша задача будет состоять в том, чтобы определить, как данные свести к одному из типов и представить в виде набора чисел.

Обычно выбор модели под тип данных выглядит следующим образом:

* **Табличный**  — классические ML-модели либо полносвязные NN;
* **Остальные типы данных** — глубокие нейронные сети.

В разных типах данных количество связей между элементами разное и зависит только от типа этих данных. **Важно НЕ количество элементов, а СВЯЗИ между ними.**


<center><img src ="https://ml.gan4x4.ru/msu/dev-2.2/L01/out/connectivity_of_data_types.png" width="700" ></center>

**Связность** — степень взаимного влияния между соседними элементами.

Например, в таблице, в которой есть определенные параметры (рост, вес), данные между собой связаны, но порядок столбцов значения не имеет. Если мы поменяем столбцы местами, то не потеряем никакой важной информации. Такие данные можно представить в виде вектора, но порядок элементов в нем не важен.

При работе с изображениями нам становится важно, как связаны между собой пиксели и по горизонтали, и по вертикали. При добавлении цвета появляются 3 RGB-канала, и значения в каждом канале также связаны между собой. Эту связь нельзя терять, если мы хотим корректно извлечь максимум информации из данных. Соответственно, если дано цветное изображение, то у нас есть уже три измерения, в которых мы должны эти связи учитывать.

## Работа с изображениями

В машинном обучении изображения — это ключевой тип данных, особенно в задачах компьютерного зрения. Но что же такое изображение? По сути, это всего лишь таблица чисел, где каждый элемент таблицы представляет собой интенсивность цвета в соответствующем пикселе.  

Изображения бывают в разных форматах:  
- **Одноканальные изображения** (Grayscale) содержат только один канал, где значение пикселя отражает яркость (от 0 до 255 для 8-битных изображений).  
- **Трёхканальные изображения** (RGB) содержат три канала: красный (R), зелёный (G) и синий (B). Каждый пиксель — это комбинация значений трёх каналов.
- **Четырехканальные изображения** (RGBA): RGB + канал прозрачности (Alpha).
- **Многоканальные данные**: Например, спутниковые снимки могут содержать десятки каналов, включая инфракрасный спектр, температуру или высоту.

Стандартный порядок для большинства изображений — это RGB (Red, Green, Blue). Однако есть случаи, когда каналы идут в другом порядке (например, BGR — синий, зелёный, красный). Например, это часто используется в библиотеке OpenCV. Полезно проверить в каком формате вы используете изображения и в каком формате их ожидают библиотеки с которыми вы работаете.

Также и порядок каналов может отличаться, каналы могут быть вначале, а могут быть в конце, для трехканальных изображений размером 100х100:
```
C, H, W (Channels, Height, Width): (3, 100, 100) - PyTorch
H, W, C (Height, Width, Channels): (100, 100, 3) - TensorFlow, OpenCV, Pillow (PIL), Matplotlib
```

Также существуют разные форматы изображений, например:

- **PNG:** Формат без потерь, поддерживает прозрачность, подходит для точного представления данных.  
- **JPEG:** Сжатие с потерями, уменьшает размер файлов, широко используется для фотографий.  
- **BMP:** Без сжатия, сохраняет исходные данные, но имеет большой размер файлов.  
- **TIFF:** Гибкий формат с поддержкой многоканальных данных и сжатия без потерь, часто используется в научных задачах.  
- **WEBP:** Эффективное сжатие с поддержкой прозрачности, экономит место, но имеет ограниченную совместимость.  
- **RAW:** Необработанные данные с камеры, сохраняет максимальное качество, но требует предварительной обработки.  

Мы можем менять изображения, например нам не нужна информация о цвете, тогда можно сделать все изображения черно-белыми (одноканальными), это можно сделать разными способами, воспользуемся базовым:

````python
image = Image.open("/content/capybara_image.jpg")
grayscale_image = image.convert("L")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Цветное изображение")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(grayscale_image, cmap="gray")
plt.title("Чёрно-белое изображение")
plt.axis("off")

plt.tight_layout()
plt.show()
````

Или поменять размер:

````python
image = Image.open("/content/capybara_image.jpg")

new_size = (100, 100)
resized_image = image.resize(new_size)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title(f"Оригинальный размер: {image.size}")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(resized_image)
plt.title(f"Изменённый размер: {new_size}")
plt.axis("off")

plt.tight_layout()
plt.show()
````
