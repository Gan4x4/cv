# How Sequence Reduction Preserves Output Size

The key to understanding why the output size remains $N \times C$ is that only the **Keys ($K$)** and **Values ($V$)** are reduced. The **Queries ($Q$)** remain at full size, and the Query dimension dictates the final output shape.

### Step 1: The Input
We start with the sequence $S$ (the flattened feature map).

$$
S \in \mathbb{R}^{N \times C}
$$

### Step 2: Generate Queries (Full Size)
The Queries are generated directly from the input without reduction.

$$
Q = S \cdot W_Q
$$



$$
\text{Shape of } Q: (N \times C)
$$

### Step 3: Generate Keys & Values (Reduced)
The input $S$ is reshaped (merging $R$ tokens) and linearly projected to reduce the sequence length by factor $R$.

$$
K = \text{Linear}(\text{Reshape}(S))
$$



$$
\text{Shape of } K, V: \left(\frac{N}{R} \times C\right)
$$

### Step 4: The Attention Map (Rectangular)
We calculate the dot product between Full $Q$ and Reduced $K^T$.

$$
\text{Attn} = \text{Softmax}\left( Q \cdot K^T \right)
$$



$$
\text{Dimensions: } \underbrace{(N \times C)}_{Q} \cdot \underbrace{\left(C \times \frac{N}{R}\right)}_{K^T} \rightarrow \left( N \times \frac{N}{R} \right)
$$

*Note: The result is a rectangular attention map, not a square one.*

### Step 5: The Final Output (Restored)
We multiply the rectangular Attention Map by the Reduced Values ($V$). The inner dimension $\frac{N}{R}$ cancels out.

$$
\text{Output} = \text{Attn} \cdot V
$$



$$
\text{Dimensions: } \underbrace{\left( N \times \frac{N}{R} \right)}_{\text{Attn}} \cdot \underbrace{\left( \frac{N}{R} \times C \right)}_{V} \rightarrow (N \times C)
$$

**Result:** The output is $N \times C$, matching the original input size.
